```python
page=requests.get("https://companiesmarketcap.com/e-commerce/largest-e-commerce-companies-by-market-cap/")
page
```

```python
Soup=BeautifulSoup(page.content)
Soup
```

```python
print(Soup.prettify())
```

```python
companyname=Soup.find_all('div',class_="company-name")
all_data=[]
for i in companyname:
    all_data.append(i.text)
all_data                    
```

```python
market=Soup.find_all('td',class_="td-right")
all_data1=[]
for i in market:
    all_data1.append(i.text)
all_data1
```

```python
marketcapital=market[1:len(all_data1):3]
cap=[]
for i in marketcapital:
    cap.append(i.text)
cap
```

```python
marketprice=market[2:len(all_data1):3]
price=[]
for i in marketprice:
    price.append(i.text)
price
```

```python
country=Soup.find_all('span',class_="responsive-hidden")
countrynames=[]
for i in country:
    countrynames.append(i.text)
countrynames
```
